```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

val = {}
val['64 x 64'] = [9.89, 10.31, 91.61, 92.46, 92.54]
val['128 x 128'] = [9.91, 10.09, 10.12, 92.56, 92.51]
val['256 x 256'] = [9.59, 9.53, 10.15, 90.43, 92.6]
val['512 x 512'] = [10.33, 9.88, 10.1, 52.64, 92.61]
val['Inf x Inf'] = [10.03, 10.07, 10.01, 67.05, 92.61]

x_tag = val.keys()
legend_tag = ["ADC 4-bit", "ADC 5-bit", "ADC 6-bit", "ADC 8-bit", "ADC 32-bit"]
dict_val = {name: v for name, v in zip(legend_tag, val.values())}
df = pd.DataFrame(dict_val, index = x_tag)
print(df)
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 0.15

index = np.arange(5)
colors = ['red', 'blue', 'green', 'salmon', 'indigo']

for i, (name, color) in enumerate(zip(df.keys(), colors)):
    plt.bar(index + i * bar_width, df[name], bar_width, alpha=0.4, color=color, label=name)
# b1 = plt.bar(index + 0 * bar_width, df['64 x 64'], bar_width, alpha=0.4, color='red', label='64 x 64')
# b2 = plt.bar(index + 1 * bar_width, df['128 x 128'], bar_width, alpha=0.4, color='blue', label='shop B')
# b3 = plt.bar(index + 2 * bar_width, df['ADC 6-bit'], bar_width, alpha=0.4, color='green', label='shop C')
# b4 = plt.bar(index + 3 * bar_width, df['ADC 8-bit'], bar_width, alpha=0.4, color='salmon', label='shop C')
# b5 = plt.bar(index + 4 * bar_width, df['ADC 32-bit'], bar_width, alpha=0.4, color='indigo', label='shop C')

# initial point , end point
plt.xticks(np.arange(2*bar_width, 5 + bar_width, 1), x_tag)

plt.xlabel('year', size = 13)
plt.ylabel('revenue', size = 13)
plt.legend()
plt.show()
```
